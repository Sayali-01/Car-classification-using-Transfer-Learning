```python
import tensorflow as tf
from tensorflow import keras

import pandas as pd
import numpy as np
import PIL
from google.colab import drive
drive.mount('/content/drive')
```

```python
from keras import layers
from keras.models import load_model
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import image_dataset_from_directory
```

```python
import matplotlib.pyplot as plt
```

```python
test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    '/content/drive/MyDrive/standford_dt/car_test/cars_test',
    seed=42,
    batch_size=50,
)
class_names = test_ds.class_names
```

```python
test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    '/content/drive/MyDrive/standford_dt/car_test/cars_test',
    seed=42,
    image_size = (224, 224),
    batch_size=50,
)
```

```python
print(" VGG16 Model")
car_path =  ["/content/drive/MyDrive/standford_dt/car_test/cars_test/00004.jpg"]
 #           "/content/drive/MyDrive/standford_dt/car_test/cars_test/audis6/001618.jpg",
  #          "/content/drive/MyDrive/standford_dt/car_test/cars_test/jeepwrangler/012008.jpg",
   #         "/content/drive/MyDrive/standford_dt/car_test/cars_test/teslamodel/015226.jpg"]
image_size = (224, 224)
model = load_model('/content/drive/MyDrive/standford_dt/vgg16')

for car in car_path:

  image = load_img(car, target_size=image_size)
  image = img_to_array(image)
  image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
  temp = plt.imread(car)
  plt.imshow(temp)
  plt.show()
  predictions = model.predict(image)
  score = tf.nn.softmax(predictions[0])
  
  print("CLASS: ", class_names[np.argmax(score)])
  print("Percent Confidence:" , np.max(score)*100)

 
```
